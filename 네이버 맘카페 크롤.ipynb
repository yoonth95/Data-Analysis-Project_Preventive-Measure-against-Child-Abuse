# 최종 결과물

```python
import requests
from bs4 import BeautifulSoup as bs
from datetime import datetime
import pandas as pd
import sys
import os
from selenium import webdriver
from webdriver_manager.chrome import ChromeDriverManager
import time
from selenium.webdriver.common.keys import Keys
import pyperclip
from selenium.webdriver.chrome.options import Options



driver=webdriver.Chrome(ChromeDriverManager().install())
driver.set_window_size(1400, 800)       # chrome 창을 원하는 가로폭과 세로폭으로 저정합니다.

url = 'https://cafe.naver.com/imsanbu'  # 맘카페 바로 들어가기
driver.get(url)
keys = Keys()

html=driver.page_source
soup=bs(html, 'html.parser')


############################################# 로그인
time.sleep(1)
driver.find_element_by_xpath('//*[@id="gnb_login_button"]').click()


tag_id = driver.find_element_by_name('id')
tag_pw = driver.find_element_by_name('pw')

tag_id.click()
pyperclip.copy('본인 아이디')          # 클립보드에 내용 복사
tag_id.send_keys(keys.CONTROL, 'v')     # id 입력란에 붙여넣기 (ctrl +v)

tag_pw.click()
pyperclip.copy('본인 비밀번호')
tag_pw.send_keys(keys.CONTROL, 'v')
time.sleep(1)

driver.find_element_by_xpath('//*[@id="log.login"]').click()    # 로그인 버튼 클릭
time.sleep(1)
# driver.find_element_by_xpath('//*[@id="new.save"]').click()
# time.sleep(2)


############################################## 키워드 입력

keyword = driver.find_element_by_name("query")
keyword.send_keys('어린이집 cctv')               ### 이 부분 키워드 바꾸시면 됩니다.

time.sleep(1)
driver.find_element_by_xpath('//*[@id="info-search"]/form/button').click()
time.sleep(1)

################################################## iframe 페이지로 전환
driver.switch_to.frame("cafe_main")
r = driver.page_source
soup = bs(r, "html.parser")

a = soup.find('div', class_='article-board result-board m-tcol-c').find_all('a', class_='article')


################################################## 내용 크롤
textlist = []
for temp in range(1, len(a)+1):
    try:
        driver.find_element_by_xpath('//*[@id="main-area"]/div[5]/table/tbody/tr['+str(temp)+']/td[1]/div[2]/div/a[1]').click()
        time.sleep(1)

        r = driver.page_source
        soup = bs(r, "html.parser")
        p = soup.find('div', class_='se-module se-module-text').find_all('p')

        txt = ''
        for i in p:
            if i.text == '\u200b':
                pass
            else:
                txt = txt + i.text
        textlist.append(txt)       

        time.sleep(1)
        driver.back()   ### 뒤로가기
        driver.switch_to.frame("cafe_main")
        time.sleep(1)
    except:
        pass

print(textlist)

driver.close()
```

```python
df = pd.DataFrame(textlist)
df.to_csv('./어린이집 cctv.csv')
```
